# Matplotlib

## Membuat bar chart

```python
import matplotlib.pyplot as plt

cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta',
          'Surakarta','Surabaya', 'Medan', 'Makassar')

populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)

plt.bar(x = cities, height = populations)

plt.show()
```

```python
plt.bar(x = cities, height = populations)

plt.xticks(rotation = 45)

plt.show()
```

```python
plt.barh(y = cities, width = populations)

plt.show()
```

Mengurutkan data

```python
import pandas as pd

df = pd.DataFrame({
    'Cities': cities,
    'Population': populations,
})

df.sort_values(by = "Population", inplace = True)

plt.barh(y = df["Cities"], width = df["Population"])
plt.show()
```

Memeberikan keterangan pada bar chart

```python
plt.barh(y = df["Cities"], width = df["Population"])
plt.xlabel("Population (Millions)")
plt.title("Population of Cities in Indonesia")
plt.show()
```

## Membuat pie chart

```python
import matplotlib.pyplot as plt

Anime = ("Sousou no Frieren", "Ookami to Koushinryou", "Koe no Katachi", "Kimi no Na wa")
votes = ( 811317, 177489, 2359488, 2772534)
colors = ('lime', 'red', 'lightblue', 'lavender')
explode = (0, 0, 0, 0.1)

plt.pie(
    x = votes,
    labels = Anime,
    autopct = '%1.1f%%',
    colors = colors,
    explode = explode
)

plt.show()
```

Versi Donut Plot

```python
plt.pie(
    x = votes,
    labels = Anime,
    colors = colors,
    wedgeprops = {'width': 0.4}
)

plt.show()
```

## Membuat histogram

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(15, 5, 250)

plt.hist(x = x, bins = 15)

plt.show()
```

# Seaborn

## Membuat bar chart

```python
import seaborn as sns

sns.barplot(y = df["Cities"], x = df["Population"], orient = 'h', color = 'indigo')

plt.xlabel("Population (Millions)")
plt.title("Population of Cities in Indonesia")
plt.show()
```

## Membuat histogram

```python
import seaborn as sns

sns.histplot(x = x, bins = 15, kde = True)

plt.show
```

## Membuat box plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

x = np.random.normal(15, 5, 250)
sns.boxplot(x)

plt.show()
```
