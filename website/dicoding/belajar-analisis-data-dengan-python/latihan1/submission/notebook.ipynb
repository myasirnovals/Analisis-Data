Deklarasi Variabel:

```python
datetime_order_items = ['shipping_limit_date']

datetime_order_reviews = [
    'review_creation_date',
    'review_answer_timestamp'
]

datetime_orders = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
```

Kode Program:

```python
for column in datetime_order_items:
    order_items_df[column] = pd.to_datetime(order_items_df[column])

for column in datetime_order_reviews:
    order_reviews_df[column] = pd.to_datetime(order_reviews_df[column])

for column in datetime_orders:
    orders_df[column] = pd.to_datetime(orders_df[column])
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Produk apa yang memiliki penjualan tertinggi dan terendah?

```python
sum_order_items_df = all_data.groupby('product_category_name_english')['product_id'].count().reset_index()
sum_order_items_df = sum_order_items_df.rename(columns={'product_id': 'product'})
sum_order_items_df = sum_order_items_df.sort_values(by='product', ascending=False)
sum_order_items_df = sum_order_items_df.head(10)

sum_order_items_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ['#72BCD4', '#D3D3D3', '#D3D3D3', '#D3D3D3', '#D3D3D3']

sns.barplot(
    x='product',
    y='product_category_name_english',
    data=sum_order_items_df.head(5),
    palette=colors,
    ax=ax[0]
)
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title('Produk dengan penjualan tertinggi', loc='center', fontsize=18)
ax[0].tick_params(axis='y', labelsize=15)

sns.barplot(
    x='product',
    y='product_category_name_english',
    data=sum_order_items_df.sort_values(by='product', ascending=True).head(5),
    palette=colors,
    ax=ax[1]
)
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position('right')
ax[1].yaxis.tick_right()
ax[1].set_title('Produk dengan penjualan terendah', loc='center', fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

plt.suptitle('Produk yang memiliki penjualan tinggi dan rendah', fontsize=20)
plt.show()
```

### Pertanyaan 2: Berapa besar pengeluaran yang telah dihabiskan oleh pelanggan dalam beberapa bulan terakhir?

```python
monthly_spend_df = all_data.resample(rule='M', on='order_approved_at').agg({
    'payment_value': 'sum'
})

monthly_spend_df.index = monthly_spend_df.index.strftime('%B')
monthly_spend_df = monthly_spend_df.reset_index()

monthly_spend_df.rename(columns={
    'payment_value': 'total_spend'
}, inplace=True)

monthly_spend_df.head()
```

```python
monthly_spend_df = monthly_spend_df.sort_values('total_spend').drop_duplicates('order_approved_at', keep='last')

monthly_spend_df.head()
```

```python
monthly_spend_df.sort_values(by='total_spend')
```

```python
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

monthly_spend_df['month_numeric'] = monthly_spend_df['order_approved_at'].map(month_mapping)
monthly_spend_df = monthly_spend_df.sort_values('month_numeric')
monthly_spend_df = monthly_spend_df.drop('month_numeric', axis=1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_spend_df['order_approved_at'],
    monthly_spend_df['total_spend'],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)

plt.title('Total Customer Spend Money per Month (2018)', loc='center', fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 3: Bagaimana perkembangan penjualan di platform E-Commerce seiring berjalannya waktu?

```python
monthly_df = all_data.resample(rule='M', on='order_approved_at').agg({
    'order_id': 'nunique',
})

monthly_df.index = monthly_df.index.strftime('%B')
monthly_df = monthly_df.reset_index()

monthly_df.rename(columns={
    'order_id': 'order_count'
}, inplace=True)

monthly_df.head()
```

```python
monthly_df = monthly_df.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')

monthly_df.head()
```

```python
monthly_df.sort_values(by='order_count')
```

```python
monthly_df['month_numeric'] = monthly_df['order_approved_at'].map(month_mapping)
monthly_df = monthly_df.sort_values('month_numeric')
monthly_df = monthly_df.drop('month_numeric', axis=1)
```

```python
plt.figure(figsize=(10, 5))

plt.plot(
    monthly_df['order_approved_at'],
    monthly_df['order_count'],
    marker='o',
    linewidth=2,
    color='#72BCD4'
)

plt.title('Number of Orders per Month (2018)', loc='center', fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

### Pertanyaan 4: Seberapa puas pelanggan terhadap layanan kami?

```python
review_scores = all_data['review_score'].value_counts().sort_values(ascending=True)

most_common_score = review_scores.idxmax()

plt.figure(figsize=(10, 5))

sns.barplot(
    x=review_scores.index,
    y=review_scores.values,
    order=review_scores.index,
    palette=['#72BCD4' if score == most_common_score else '#D3D3D3' for score in review_scores.index]
)

plt.title('Rating by Customer', fontsize=15)
plt.xlabel(None)
plt.ylabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### Pertanyaan 5: Seperti apa profil demografis pelanggan dan apakah terdapat perbedaan preferensi pembelian di antara mereka?

#### Berdasarkan Negara (State)

```python
bystate_df = all_data.groupby(by='customer_state').customer_id.nunique().reset_index()

bystate_df.rename(columns={
    'customer_id': 'customer_count'
}, inplace=True)

bystate_df.head()
```

```python
plt.figure(figsize=(12, 8))

most_common_state = bystate_df.loc[bystate_df['customer_count'].idxmax(), 'customer_state']

bystate_df = bystate_df.sort_values(by='customer_count', ascending=False)

sns.barplot(
    x='customer_count',
    y='customer_state',
    data=bystate_df,
    palette=['#72BCD4' if state == most_common_state else '#D3D3D3' for state in bystate_df['customer_state']]
)

plt.title('Number of Customer by States', loc='center', fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

#### Berdasarkan Kota (City)

```python
bycity_df = all_data['customer_city'].value_counts().head(10)

plt.figure(figsize=(10, 5))

most_common_city = bycity_df.idxmax()

bycity_df = bycity_df.sort_values(ascending=False)

sns.barplot(
    x=bycity_df.index,
    y=bycity_df.values,
    palette=['#72BCD4' if city == most_common_city else '#D3D3D3' for city in bycity_df.index]
)

plt.title('Number of Customer by City', fontsize=15)
plt.xlabel(None)
plt.ylabel(None)
plt.xticks(rotation=45, fontsize=10)
plt.show()
```
